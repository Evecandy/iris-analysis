```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
from google.colab import files
import os

# Use files.upload() to get the file from your local computer
# This will open a file picker dialog
# Ensure this cell is run FIRST to load the DataFrame
uploaded = files.upload()

# Check if any file was uploaded
if uploaded:
    # Get the name of the first uploaded file (assuming only one is uploaded)
    file_name = list(uploaded.keys())[0]
    print(f"Uploaded file: {file_name}")

    # Load the dataset from the uploaded file
    try:
        df = pd.read_csv(file_name)
        print("Dataset loaded successfully.")

        # Verify the first few rows to confirm successful loading
        print(df.head())

        # Now proceed with the plotting code
        # 🔹 Bar Plot
        if 'species' in df.columns:
            sns.countplot(x=df['species'], palette='viridis')
            plt.xlabel("Species Type")
            plt.ylabel("Count")
            plt.title("Distribution of Iris Species")
            plt.savefig("bar_plot_species_distribution.png", dpi=300)
            plt.show()
        else:
            print("Error: 'species' column not found for Bar Plot.")


        # 🔹 Line Chart
        # Check for required columns before plotting
        if all(col in df.columns for col in ['petal_length', 'species']):
            sns.lineplot(x=df.index, y=df['petal_length'], hue=df['species'])
            plt.xlabel("Sample Index")
            plt.ylabel("Petal Length")
            plt.title("Petal Length Variation Across Samples")
            plt.savefig("line_chart_petal_length.png", dpi=300)
            plt.show()
        else:
            print("Error: Required columns ('petal_length', 'species') not found for Line Chart.")


        # 🔹 Scatter Plot (Sepal vs Petal Length)
        # Check for required columns before plotting
        if all(col in df.columns for col in ['sepal_length', 'petal_length', 'species']):
            sns.scatterplot(x=df['sepal_length'], y=df['petal_length'], hue=df['species'], palette='coolwarm')
            plt.xlabel("Sepal Length")
            plt.ylabel("Petal Length")
            plt.title("Sepal Length vs Petal Length Across Species")
            plt.savefig("scatter_plot_sepal_vs_petal.png", dpi=300)
            plt.show()
        else:
             print("Error: Required columns ('sepal_length', 'petal_length', 'species') not found for Scatter Plot.")


        # 🔹 Boxplot
        # Assuming boxplot is for numerical columns, no specific column check needed unless excluding non-numeric
        plt.figure() # Create a new figure for the boxplot
        sns.boxplot(data=df)
        plt.title("Boxplot of Iris Features")
        plt.savefig("boxplot_iris_features.png", dpi=300)
        plt.show()


        # 🔹 Heatmap (Feature Correlation)
        # Check for numerical columns before plotting the heatmap
        numerical_df = df.select_dtypes(include=['number'])
        if not numerical_df.empty:
            plt.figure() # Create a new figure for the heatmap
            sns.heatmap(numerical_df.corr(), annot=True, cmap="coolwarm")
            plt.title("Feature Correlation Heatmap")
            plt.savefig("heatmap_correlation.png", dpi=300)
            plt.show()
        else:
            print("Error: No numerical columns found for Heatmap.")


    except Exception as e:
        print(f"Error loading the dataset or creating the plot: {e}")

else:
    print("No file was uploaded. Please upload the dataset file.")
```
